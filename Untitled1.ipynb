```python
#匹配流派
import pandas as pd
import numpy as np
from numpy import *
data_by_artist = pd.read_csv('data_by_artist.csv',sep=",")
data_by_year = pd.read_csv('data_by_year.csv',sep=",")
full_music_data = pd.read_csv('full_music_data.csv',sep=",")
influence_data = pd.read_csv('influence_data.csv',sep=',')

genredict=dict(zip(influence_data["influencer_id"],influence_data["influencer_main_genre"]))
genredict2=dict(zip(influence_data["follower_id"],influence_data["follower_main_genre"]))
genredict.update(genredict2)
```

```python
#求流派内的马氏距离
genreset=[]
mszn=[]
genrelist=list(set(genr))#20个流派
#weight=model.feature_importances_
for i in genrelist:#遍历流派
    a=data_by_artist[data_by_artist['genre'].isin([i])].iloc[:,2:-2]
    b=np.cov(a.T)
    if np.linalg.det(b)==0:
        for q in range(b.shape[0]):
            b[q,q]=b[q,q]+1e-6
    invb=np.linalg.inv(b)#计算协方差矩阵的逆
    ms=[]
    for j in a.index:
        for k in a.index:
            c=np.array(a.loc[j])*weight
            d=np.array(a.loc[k])
            tp=c-d
            ms.append(np.sqrt(dot(dot(tp,invb),tp.T)))
    mszn.append(max(ms))
mszz=pd.DataFrame(mszn.T,columns=genrelist)
mszz
```

```python
import numpy as np#组间距离
from numpy import *
mszj=[]
for i in genrelist:#遍历流派
    dataa=data_by_artist[data_by_artist['genre'].isin([i])].iloc[:,2:-2]
    for j in genrelist:
        datab=data_by_artist[data_by_artist['genre'].isin([j])].iloc[:,2:-2]
        dataset=pd.merge(dataa,datab)
        d=np.cov(dataset.T)
        if np.linalg.det(d)==0:
            for q in range(d.shape[0]):
                d[q,q]=d[q,q]+1e-6
        invb=np.linalg.inv(d)#两个流派合并数据的协方差矩阵
        for k in dataa.index:
            for q in datab.index:
                c=np.array(dataa.loc[k])*weight
                d=np.array(datab.loc[q])
                tp=c-d
                mszj.append(np.sqrt(dot(dot(tp,invb),tp.T)))
mszj
```

```python
#计算入度
influence = influence_data.copy()
genre_father_id = {}
influence['root_id'] = influence['influencer_id']

for genre in genrelist:
    # 取出某一类的所有
    df = influence.loc[influence['follower_main_genre']==genre, :]
    
#     df = df.loc[df['influencer_main_genre']==genre, :]
    # 父节点初始化
    fathers_id={}
    for i in df.index:
        if not fathers_id.get(df.loc[i, 'follower_id']):
            fathers_id[df.loc[i, 'follower_id']] = set([df.loc[i, 'follower_id']])
        else:
            fathers_id[df.loc[i, 'follower_id']].add(df.loc[i, 'influencer_id'])
        if not fathers_id.get(df.loc[i, 'influencer_id']):
            fathers_id[df.loc[i, 'influencer_id']] = set([df.loc[i, 'influencer_id']])
#         else:
#             fathers_id[df.loc[i, 'influencer_id']].add(df.loc[i, 'influencer_id'])
    genre_father_id[genre]=fathers_id
for i in genre_father_id.keys():
    for j in genre_father_id[i].keys():
        genre_father_id[i][j]=list(genre_father_id[i][j])
        if j in genre_father_id[i][j] and len(genre_father_id[i][j])>1:
            genre_father_id[i][j].remove(j)
genre_father_id
```

```python
#计算出度
genre_son_id = {}

for genre in genrelist:
    # 取出某一类的所有
    df = influence.loc[influence['follower_main_genre']==genre, :]
    
    # 子节点初始化
    son_id={}
    for i in df.index:
        if not son_id.get(df.loc[i, 'influencer_id']):
            son_id[df.loc[i, 'influencer_id']] = set([df.loc[i, 'influencer_id']])
        else:
            son_id[df.loc[i, 'influencer_id']].add(df.loc[i, 'follower_id'])
        if not son_id.get(df.loc[i, 'follower_id']):
            son_id[df.loc[i, 'follower_id']] = set([df.loc[i, 'follower_id']])
#         else:
#             fathers_id[df.loc[i, 'influencer_id']].add(df.loc[i, 'influencer_id'])
    genre_son_id[genre]=son_id
for i in genre_son_id.keys():
    for j in genre_son_id[i].keys():
        genre_son_id[i][j]=list(genre_son_id[i][j])
        if j in genre_son_id[i][j] and len(genre_son_id[i][j])>1:
            genre_son_id[i][j].remove(j)
genre_son_id
```
